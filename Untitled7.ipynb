```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.utils import to_categorical
```

```python
# === 1. Set Paths and Parameters ===
img_size = 224
batch_size = 32
train_dir = 'chest_xray/train'
val_dir = 'chest_xray/val'
test_dir = 'chest_xray/test'
```

```python
# === 2. Data Generators with Augmentation ===
train_aug = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)
```

```python
val_test_aug = ImageDataGenerator(rescale=1./255)
```

```python

train_generator = train_aug.flow_from_directory(train_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical')
val_generator = val_test_aug.flow_from_directory(val_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical')
test_generator = val_test_aug.flow_from_directory(test_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical', shuffle=False)
```

```python
class_names = list(train_generator.class_indices.keys())
num_classes = len(class_names)
```

```python
# === 3. Build Model ===
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(img_size, img_size, 3))
base_model.trainable = False  # Freeze all initially
```

```python
model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])
```

```python
# === 4. Compile ===
model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# === 5. Train (Phase 1 - Frozen) ===
callbacks = [
    ModelCheckpoint("best_model.keras", save_best_only=True, monitor="val_accuracy", mode="max"),
    EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)
]
```

```python
istory1 = model.fit(train_generator, validation_data=val_generator, epochs=10, callbacks=callbacks)
```

```python
# === 6. Unfreeze Top Layers and Fine-tune (Phase 2) ===
base_model.trainable = True
for layer in base_model.layers[:-4]:  # Fine-tune only top 4
    layer.trainable = False
```

```python
model.compile(optimizer=Adam(1e-5), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history2 = model.fit(train_generator, validation_data=val_generator, epochs=10, callbacks=callbacks)
```

```python
# === 7. Save Final Model ===
model.save("fine_tuned_pneumonia_model10.keras")
```

```python
# === 8. Evaluate on Test Set ===
test_generator.reset()
Y_pred = model.predict(test_generator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = test_generator.classes
y_true_onehot = to_categorical(y_true, num_classes=num_classes)

print("\n📋 Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))
```

```python

test_acc = np.sum(y_pred == y_true) / len(y_true) * 100
print(f"✅ Final Test Accuracy: {test_acc:.2f}%")
```

```python
# === 9. Accuracy and Loss Plots ===
def plot_history(history, title):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

plot_history(istory1, "Phase 1")
plot_history(history2, "Phase 2")
```

```python
# === 10. AUROC Curve ===
if num_classes > 2:
    y_true_bin = label_binarize(y_true, classes=list(range(num_classes)))
    fpr, tpr, roc_auc = {}, {}, {}

    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], Y_pred[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    plt.figure(figsize=(8, 6))
    for i, color in zip(range(num_classes), ['red', 'green', 'blue']):
        plt.plot(fpr[i], tpr[i], color=color, label=f'{class_names[i]} (AUC = {roc_auc[i]:.4f})')

    plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')
    plt.title('Multiclass AUROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
else:
    fpr, tpr, _ = roc_curve(y_true, Y_pred[:, 1])
    auc_score = roc_auc_score(y_true, Y_pred[:, 1])
    plt.figure()
    plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title('AUROC Curve')
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# === 11. Load Trained Model for Prediction ===
from tensorflow.keras.models import load_model
model = load_model("fine_tuned_pneumonia_model10.keras")
```

```python

# === 12. Predict from Path ===
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
def predict_from_path(image_path):
    img = load_img(image_path, target_size=(img_size, img_size))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)
    confidence = prediction[0][predicted_class] * 100
    print(f"\n🖼️ Image Path: {image_path}")
    print(f"✅ Predicted Class: {class_names[predicted_class]}")
    print(f"📊 Confidence: {confidence:.2f}%")
    plt.imshow(load_img(image_path))
    plt.title(f"{class_names[predicted_class]} ({confidence:.2f}%)")
    plt.axis('off')
    plt.show()

# === 13. Example Test ===
predict_from_path('chest_xray/test/NON_CHEST_XRAY/image124.png')
```
